# Matplotlib

# Scatter Plot

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,marker='.',color='red',label='data')
plt.xlabel('Country Name')
plt.ylabel('Birth rate')
plt.title('Scatter Plot Of Birth rate of each Country')
plt.grid(True)
plt.legend()
plt.show()
```

# Line Plot

```python
plt.plot(x,y,marker='*',label='data')
plt.xlabel('Country Name')
plt.ylabel('Birth rate')
plt.title('Line plot of Birth rate of country name')
plt.legend()
plt.show()
```

# Bar Plot

```python
plt.bar(x,y,color='Orange',label='data')
plt.xlabel('Country Name')
plt.ylabel('Birth rate')
plt.title('Bar plot of Birth rate of each Country')
plt.legend()
plt.show()
```

# Box Plot

```python
plt.boxplot(data)
plt.xlabel('Data')
plt.ylabel('value')
plt.title('Boxplot')
plt.grid(True)
plt.show()
```

# Heat Map

```python
plt.imshow(data,cmap='coolwarm',interpolation='nearest')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```

# Stacked Bar Plot

```python
categories = ['A','B','C','D']
values1 =[20,13,54,23]
values2=[10,7,12,5]

plt.bar(categories,values1,color='Purple',label='Data1')
plt.bar(categories,values2,color='Yellow',label='Data2')
plt.xlabel=('Categories')
plt.ylabel=('Values')
plt.title('Stacked bar plot')
plt.legend()
plt.show()
```
